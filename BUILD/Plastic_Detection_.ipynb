```python
import os
import random
import cv2
import matplotlib.pyplot as plt

# Set paths
img_dir = "/content/drive/MyDrive/Plastic_Detection Yolov8/BTS-V1-2/valid/images"
label_dir = "/content/drive/MyDrive/Plastic_Detection Yolov8/BTS-V1-2/valid/labels"

# List image files
image_files = [f for f in os.listdir(img_dir) if f.endswith(('.jpg', '.jpeg', '.png'))]

# Function to draw bounding boxes
def draw_yolo_boxes(img_path, label_path):
    img = cv2.imread(img_path)
    h, w, _ = img.shape

    # Check if label file exists
    if not os.path.exists(label_path):
        return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    with open(label_path, "r") as f:
        for line in f.readlines():
            cls, x, y, bw, bh = map(float, line.strip().split())
            # Convert YOLO format to box coordinates
            x1 = int((x - bw / 2) * w)
            y1 = int((y - bh / 2) * h)
            x2 = int((x + bw / 2) * w)
            y2 = int((y + bh / 2) * h)
            # Draw rectangle and class ID
            cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(img, f'Class {int(cls)}', (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 0), 2)

    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Show 5 random images with bounding boxes
plt.figure(figsize=(15, 10))

for i in range(5):
    filename = random.choice(image_files)
    img_path = os.path.join(img_dir, filename)
    label_path = os.path.join(label_dir, os.path.splitext(filename)[0] + ".txt")

    img_with_boxes = draw_yolo_boxes(img_path, label_path)

    plt.subplot(1, 5, i + 1)
    plt.imshow(img_with_boxes)
    plt.title(filename)
    plt.axis("off")

plt.tight_layout()
plt.show()
```
