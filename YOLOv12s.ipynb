# **YOLOv12n**

```python
# Define the YAML content as a string
yaml_content = """
# yolov12n.yaml — Nano model definition (manual)
# Lighter than YOLOv12s

path: ../datasets/coco128  # dataset root dir
train: images/train2017 # train images (relative to 'path') 128 images
val: images/val2017 # val images (relative to 'path') 128 images
test: # test images (optional)

# Classes
nc: 15  # number of classes
names: ['class1', 'class2', 'class3', 'class4', 'class5', 'class6', 'class7', 'class8', 'class9', 'class10', 'class11', 'class12', 'class13', 'class14', 'class15']  # class names

depth_multiple: 0.33
width_multiple: 0.25

backbone:
  [[-1, 1, Conv, [64, 6, 2, 2]],
   [-1, 1, Conv, [128, 3, 2]],
   [-1, 3, C2f, [128]],
   [-1, 1, Conv, [256, 3, 2]],
   [-1, 6, C2f, [256]],
   [-1, 1, Conv, [512, 3, 2]],
   [-1, 6, C2f, [512]],
   [-1, 1, Conv, [1024, 3, 2]],
   [-1, 3, C2f, [1024]],
   [-1, 1, SPPF, [1024, 5]]]

head:
  [[-1, 1, Conv, [512, 1, 1]],
   [-1, 1, nn.Upsample, [None, 2, 'nearest']],
   [[-1, 6], 1, Concat, [1]],
   [-1, 3, C2f, [512]],

   [-1, 1, Conv, [256, 1, 1]],
   [-1, 1, nn.Upsample, [None, 2, 'nearest']],
   [[-1, 4], 1, Concat, [1]],
   [-1, 3, C2f, [256]],

   [-1, 1, Conv, [128, 1, 1]],
   [-1, 1, nn.Upsample, [None, 2, 'nearest']],
   [[-1, 2], 1, Concat, [1]],
   [-1, 3, C2f, [128]],

   [[10, 13, 16], 1, Detect, [15]]  # ⚠️ Replace 15 with your actual number of classes
  ]
"""

# Specify the filename
file_name = "yolov12n.yaml"

# Write the content to the file
with open(file_name, "w") as f:
    f.write(yaml_content)

print(f"✅ Created '{file_name}'")
```

```python
from ultralytics import YOLO

# Step 1: Load YOLOv12n architecture (nano), training from scratch
model = YOLO('/content/yolov12n.yaml')  # Nano version config file

# Step 2: Train the model
results = model.train(
    data='/content/drive/MyDrive/OCEAN DATASET/data.yaml',  # COCO-style dataset
    epochs=50,
    imgsz=640,
    batch=16,
    name='yolov12n_custom_run',  # You can change this name
    pretrained=False,  # 🔴 From scratch, no pre-trained weights
    optimizer='Adam',  # Optional: same as your other experiments
    lr0=0.0001,         # Optional tuning (based on your m/m/l setup)
    weight_decay=0.0005,
    workers=2
)
```
