```python
from transformers import YolosImageProcessor, YolosForObjectDetection
from torchvision.datasets import CocoDetection
from torch.utils.data import DataLoader
from torchvision.transforms import ToTensor
import torchvision.transforms.functional as F
from PIL import Image
import torch
import os

# Load processor and model
processor = YolosImageProcessor.from_pretrained("hustvl/yolos-tiny")
model = YolosForObjectDetection.from_pretrained("hustvl/yolos-tiny")

# Dataset path
data_root = "/content/drive/MyDrive/VT"
train_dir = os.path.join(data_root, "train")
val_dir = os.path.join(data_root, "valid") # Corrected val directory name
train_json = os.path.join(data_root, "train/_annotations.coco.json") # Corrected path to train annotations
val_json = os.path.join(data_root, "valid/_annotations.coco.json") # Corrected path to validation annotations

# COCO Dataset wrapper
class CocoDetectionTransforms(CocoDetection):
    def __init__(self, img_folder, ann_file, processor):
        super().__init__(img_folder, ann_file)
        self.processor = processor

    def __getitem__(self, idx):
        img, target = super().__getitem__(idx)
        target = {"image_id": idx, "annotations": target}
        encoding = self.processor(images=img, annotations=target, return_tensors="pt")
        pixel_values = encoding["pixel_values"].squeeze()
        labels = encoding["labels"][0]
        return pixel_values, labels

train_dataset = CocoDetectionTransforms(train_dir, train_json, processor)
val_dataset = CocoDetectionTransforms(val_dir, val_json, processor)
```

```python
from torch.utils.data import DataLoader
from torch.optim import AdamW
from tqdm import tqdm

train_loader = DataLoader(train_dataset, batch_size=2, shuffle=True, collate_fn=lambda x: list(zip(*x)))
val_loader = DataLoader(val_dataset, batch_size=2, collate_fn=lambda x: list(zip(*x)))

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

optimizer = AdamW(model.parameters(), lr=5e-5, weight_decay=0.01)
```

```python
EPOCHS = 50

for epoch in range(EPOCHS):
    model.train()
    total_loss = 0

    for pixel_values, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{EPOCHS}"):
        pixel_values = torch.stack(pixel_values).to(device)
        labels = [{k: v.to(device) for k, v in t.items()} for t in labels]

        outputs = model(pixel_values=pixel_values, labels=labels)
        loss = outputs.loss

        loss.backward()
        optimizer.step()
        optimizer.zero_grad()

        total_loss += loss.item()

    avg_loss = total_loss / len(train_loader)
    print(f"Epoch {epoch+1}/{EPOCHS}, Loss: {avg_loss:.4f}")
```
