# **YOLOv12m**

```python
# STEP 1: Install YOLOv12 from GitHub
!pip install -q git+https://github.com/sunsmarterjie/yolov12.git

# STEP 2: Import the YOLO module
from ultralytics import YOLO

# STEP 3: Load the pretrained YOLOv12m model
model = YOLO('/content/yolov12m.pt')  # ✅ Make sure this path matches where you uploaded the file

# STEP 4: Train the model on your custom dataset
results = model.train(
    data='/content/drive/MyDrive/UPDATED YOLO  /data.yaml',  # ✅ Update this path if needed
    epochs=50,
    imgsz=640,
    batch=8,  # Reduced batch size to mitigate OutOfMemoryError
    optimizer='Adam',
    lr0=0.0001,
    weight_decay=0.0005,
    name='yolov12m_run1',
    pretrained=True,
    workers=2
)
```
