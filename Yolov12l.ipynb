```python
# Step 1: Install Ultralytics
!pip install ultralytics --upgrade -q

# Step 2: Import YOLO class
from ultralytics import YOLO

# Step 3: Download YOLOv8-L model file (assuming intent was YOLOv8)
!wget https://github.com/ultralytics/assets/releases/download/v8.1.0/yolov8l.pt -O yolov8l.pt

# Step 4: Load YOLOv8-L model
# Load the pre-trained YOLOv8-L model from the downloaded file.
model = YOLO('yolov8l.pt')
```

```python
results = model.train(
    data='/content/drive/MyDrive/OCEAN/data.yaml',  # ✅ Your dataset path
    epochs=50,
    imgsz=640,
    batch=8,
    optimizer='Adam',
    lr0=0.0001,
    weight_decay=0.0005,
    name='yolov12l_run1',  # 💡 Unique run name
    pretrained=True,
    workers=2
)
```
